```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
HRD_pan_cancer = pd.read_csv('../../HRD_score/data/HRD_scores_pan_cancer_annotated_v2.csv', sep=',', header = 0)
primary = HRD_pan_cancer[HRD_pan_cancer['Type'] == 'Primary']
luad = primary[primary['Project ID'] == 'TCGA-LUAD']
```

```python
sorted_df = luad.sort_values('HRD_sum', ascending=False)
```

```python
xdata = np.array(sorted_df['TAI'])
ydata = np.array(sorted_df['LST'])
zdata = np.array(sorted_df['LOH'])
hrd = np.array(sorted_df['HRD_sum'])
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
#ax.scatter3D(xdata, ydata, zdata, c = zdata)
ax.plot_trisurf(xdata, ydata, zdata,
                cmap='viridis', edgecolor='none')
ax.set_xlabel('TAI')
ax.set_ylabel('LST')
ax.set_zlabel('LOH')

```
